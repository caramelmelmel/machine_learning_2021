# Reading functions

```python
def read_data(path):
  dataset = []
  f = open(path,"r")
  training_set = f.readlines()
  for line in training_set:
    #loop backwards
    line = line.rstrip('\n')
    line = line.rpartition(' ')
    line = list(line)
    del line[1]
    if line != ['', '']:
      dataset.append(line)
  return dataset
```

```python
#test the read function
RU_training =read_data(RU_train)
```

```python
def read_test(path):
    f = open(path,"r", encoding="utf-8")
    test_data = f.read().splitlines()
    lines = []
    para = 0
    for data in test_data:
        if data != "":
            lines.append([data,para])
        else:
            para += 1
    return lines
```

```python
RU_testing = read_test(RU_test)
print(RU_testing)
```

```python
# y -> x
# Output: dictionary of form {'label', {'word1': count of word1 in this label, ...}}
# label to word individual count
def count_tag_words(training_set):
    label_to_indiv_word_count ={'O':{},
    'B-positive':{},
    'B-neutral':{},
    'B-negative':{},
    'I-positive':{},
    'I-neutral':{},
    'I-negative':{}}
    for tag_word_pair in training_set:
        #label
        label = tag_word_pair[1]
        #word
        word = tag_word_pair[0]
        if word not in label_to_indiv_word_count[label].keys():
            label_to_indiv_word_count[label][word] = 1
        else:
            label_to_indiv_word_count[label][word] += 1
    return label_to_indiv_word_count

```

```python
# count labels in training set
def count_labels(training_set):
    label_count ={'O':0,
        'B-positive':0,
        'B-neutral':0,
        'B-negative':0,
        'I-positive':0,
        'I-neutral':0,
        'I-negative':0}
    for label_word_pair in training_set:
        #print(label_word_pair)
        label = label_word_pair[1]
        label_count[label] += 1 
    return label_count
```

```python
#estimates the emission for each word and label
#output: {label:{word:count}}
def estimate_emission_parameters(counted_tags,counted_tag_words):
    all_estimations = {}
    for tag, word_count_dict in counted_tag_words.items():
        if tag not in all_estimations.keys():
            all_estimations[tag] = {}
        for word, count in word_count_dict.items():
            if word not in all_estimations[tag].keys():
                all_estimations[tag][word] = count/counted_tags[tag]
    return all_estimations
```

```python
# output
word_dict = count_tag_words(RU_training)
label_dict = count_labels(RU_training)
emission = estimate_emission_parameters(label_dict,word_dict)
print(emission)
```

```python
def check_for_unk(training,testing):
    training_set = []
    for word_label_pair in training:
        word = word_label_pair[0]
        training_set.append(word)
        
    for index,test_word in enumerate(testing):
        word_in_test = test_word[0]
        para = test_word[1]
        if word_in_test not in training_set:
            testing[index] = ['#UNK#',para]
    return testing
        
```

```python
test_with_unk = check_for_unk(RU_training,RU_testing)
print(test_with_unk)
```

# Emission Fix Implementation

```python
def modified_denominator_label(old_count_dict,k=1):
    for label,count in old_count_dict.items():
        old_count_dict[label] = count + k
    return old_count_dict
```

```python
def fix_count_emission(train_y_to_x,y_p_1_dict,test,k=1):
    word_to_prob = {}
    word_to_best_label = {}
    #emission
    for label, word_count_dict in train_y_to_x.items():
        for test_inputs in test:

            test_word = test_inputs[0]
            para_idx = test_inputs[1]
            if test_word == '#UNK#':
                e_xy = k/y_p_1_dict[label]
            else:
                if test_word in word_count_dict.keys():
                    e_xy = word_count_dict[test_word]/y_p_1_dict[label]
                else:
                    e_xy = 0

            if test_word not in word_to_prob.keys():
                word_to_prob[test_word] = e_xy
                word_to_best_label[test_word] = label
            else:
                if word_to_prob[test_word] < e_xy:
                    word_to_prob[test_word] = e_xy
                    word_to_best_label[test_word] = label

    return word_to_best_label
        
    
```
